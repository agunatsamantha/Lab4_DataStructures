```python
# Required Inputs for All Exercises
SEED_NUM = 4
STUDENT_MAJOR = "BSECE"           
FAVORITE_ARTIST = "Cup of Joe"          
LOCAL_HAZARD = "TYPHOON"         
CONTROL_NUM = max(3, SEED_NUM)    

print(f"--- Global Variables Initialized ---")
print(f"SEED_NUM: {SEED_NUM}")
print(f"CONTROL_NUM: {CONTROL_NUM}")
```

```python
# Initializing the official lineup 

festival = {"Ben&Ben", "SB19", "Bini", "Eraserheads", "FAVORITE_ARTIST", "Zild",
           f"Indie Artist {CONTROL_NUM}" }

# Initalizing two User Profiles as sets

user_a = {"Ben&Ben", "Bini", "Maki", "Dionela", "FAVORITE_ARTIST"}
user_b = {"SB19", "Eraserheads", "Zild", f"Indie Artist {CONTROL_NUM}",
          "Parokya ni Edgar"}

# Data expansion
# CONTOL_NUM = new artists

new_artists = {"Fitterkarma", "Itchyworms", "Apo Hiking Society", "Kamikazee" }
festival.update(new_artists)

# User A

intersection_a = len(user_a & festival)
union_a = len(user_a | festival)
jaccard_a = (intersection_a / union_a) * 100

# User B

intersection_b = len(user_b & festival)
union_b = len(user_b | festival)
jaccard_b = (intersection_b / union_a) * 100

# Dealbreakers of User A

dealbreakers_a = user_a - festival

print("=== EXERCISE 2: JACCARD SIMILARITY ENGINE ===")
print(f"Total Artists in Expanded Festival: {len(festival)}")
print(f"User A and Festival Length of Intersection: {intersection_a} ")
print(f"User A and Festival Length of Union: {union_a}")
print(f"User A Jaccard Similarity: {jaccard_a}")
print(f"User B Jaccard Similarity: {jaccard_b}")
print(f"Dealbreaker Artists of User A: {dealbreakers_a} ")


```
