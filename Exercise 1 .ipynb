```python
# Required Inputs for All Exercises
SEED_NUM = 4
STUDENT_MAJOR = "BSECE"           
FAVORITE_ARTIST = "Cup of Joe"          
LOCAL_HAZARD = "TYPHOON"         
CONTROL_NUM = max(3, SEED_NUM)    

print(f"--- Global Variables Initialized ---")
print(f"SEED_NUM: {SEED_NUM}")
print(f"CONTROL_NUM: {CONTROL_NUM}")
```

```python
from collections import defaultdict, Counter

# Initialize the simulated stream
stream = [
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", STUDENT_MAJOR),
    ("TUP Taguig", "BSECE"),
    ("TUP Manila", "BSECE")
]

# Data Expansion: Add exactly (CONTROL_NUM + 3) = 6 custom tuples
custom_applications = [
    ("TUP Manila", "BSCS"),
    ("TUP Cavite", "BSECE"),
    ("TUP Manila", "BSIT"),
    ("TUP Taguig", "BSIT"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", "BSME")
]
stream.extend(custom_applications)

# Algorithm Phase 1 (Aggregation)
campus_data = defaultdict(list)
for campus, program in stream:
    campus_data[campus].append(program)

# Algorithm Phase 2 (Frequency Analysis)
campus_frequencies = {}
for campus, programs in campus_data.items():
    campus_frequencies[campus] = Counter(programs)

# Extract top program for Manila
manila_top_program = campus_frequencies["TUP Manila"].most_common(1)[0]

# Outputs for Assessment Table
print("=== EXERCISE 1 ASSESSMENT DATA ===")
print(f"CONTROL_NUM Used: {CONTROL_NUM}")
print(f"STUDENT_MAJOR Used: {STUDENT_MAJOR}")
print(f"Expanded Stream Length: {len(stream)}")
print(f"Total Applications Processed for TUP Manila: {len(campus_data['TUP Manila'])}")
print(f"Top Requested Program at TUP Manila: {manila_top_program[0]}")
print(f"Exact Frequency of Top Program: {manila_top_program[1]}")
```
