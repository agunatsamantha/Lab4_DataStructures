```python
# Required Inputs for All Exercises
SEED_NUM = 4
STUDENT_MAJOR = "BSECE"           
FAVORITE_ARTIST = "Cup of Joe"          
LOCAL_HAZARD = "TYPHOON"         
CONTROL_NUM = max(3, SEED_NUM)    

print(f"--- Global Variables Initialized ---")
print(f"SEED_NUM: {SEED_NUM}")
print(f"CONTROL_NUM: {CONTROL_NUM}")
```

```python
from collections import deque, Counter

# Initialize deque with strict memory limit
alert_window = deque(maxlen=5)

# Simulate burst of incoming alerts
burst = [
    (CONTROL_NUM, "WEATHER"),
    (CONTROL_NUM + 1, "TRAFFIC"),
    (CONTROL_NUM + 2, "WEATHER"),
    (CONTROL_NUM + 3, "WEATHER"),
    (CONTROL_NUM + 4, "LOCAL_HAZARD"),
    (CONTROL_NUM + 5, "WEATHER"),
    (CONTROL_NUM + 6, "TRAFFIC")
]

# Data Expansion: Append exactly (CONTROL_NUM + 2) = 6 custom alerts
# Timestamps must continue sequentially from where base list left off (CONTROL_NUM + 6 = 10)
last_timestamp = CONTROL_NUM + 6
custom_alerts = [
    (last_timestamp + 1, "FIRE"),
    (last_timestamp + 2, "WEATHER"),
    (last_timestamp + 3, "LOCAL_HAZARD"),
    (last_timestamp + 4, "TRAFFIC"),
    (last_timestamp + 5, "EARTHQUAKE"),
    (last_timestamp + 6, "EARTHQUAKE")
]
burst.extend(custom_alerts)

# Algorithm (The Sliding Window)
for alert in burst:
    timestamp, category = alert
    
    # Extract just the categories currently in the deque
    current_categories = [item[1] for item in alert_window]
    
    # Pass to a Counter
    counts = Counter(current_categories)
    
    # Spam Rule: Reject if category exists 2 or more times
    if counts [category] >= 2:
        print(f"REJECTED: {timestamp} - {category} (Spam detected)")
    else:
        alert_window.append((timestamp, category))
        print(f"ACCEPTED: {timestamp} - {category}")

# Sum up all Timestamp values of successfully stored alerts
total_timestamp_sum = sum(alert[0] for alert in alert_window)

# Outputs for Assessment
print("\n=== EXERCISE 3 FINAL STATE ===")
print("\nFinal Deque State:", list(alert_window))
print(f"\nSum of Final Timestamps: {total_timestamp_sum}")
```
